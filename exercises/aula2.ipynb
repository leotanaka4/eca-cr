```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Definindo uma função para calcular o erro relativo
def erro_relativo(theta):
    return abs(math.sin(theta) - theta) / abs(math.sin(theta))
```

```python
# Gerando valores de theta de 0 até pi/2
thetas = np.linspace(0, math.pi/2, 400)
sins = [math.sin(theta) for theta in thetas]
linear_approximations = thetas
relative_errors = [erro_relativo(theta) for theta in thetas]
```

```python
# Plotando sin(theta) e sua aproximação linear
plt.figure(figsize=(10, 6))
plt.plot(thetas, sins, label='sin(θ)', color='blue')
plt.plot(thetas, linear_approximations, label='Aproximação Linear θ', linestyle='--', color='red')
plt.axvline(x=math.pi/4, color='green', linestyle='--', label='θ = π/4')
plt.axvline(x=math.pi/2, color='purple', linestyle='--', label='θ = π/2')
plt.title('sin(θ) e sua Aproximação Linear')
plt.xlabel('θ')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotando o erro relativo
plt.figure(figsize=(10, 6))
plt.plot(thetas, relative_errors, label='Erro Relativo', color='orange')
plt.axvline(x=math.pi/4, color='green', linestyle='--', label='θ = π/4')
plt.axvline(x=math.pi/2, color='purple', linestyle='--', label='θ = π/2')
plt.title('Erro Relativo')
plt.xlabel('θ')
plt.ylabel('Erro Relativo')
plt.legend()
plt.grid(True)
plt.show()
```
